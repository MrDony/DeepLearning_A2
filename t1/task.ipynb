```python
import json

# Define the path to your JSON file
file_path = '1.7_history.json'

# Initialize an empty list to store the parsed JSON objects
json_data = []

# Open the file and read its contents
with open(file_path, 'r') as file:
    # Load the JSON data from the file into a list
    json_data = json.load(file)

# Now, json_data contains a list of dictionaries, each representing a JSON object
# You can access the elements in the list as needed
for obj in json_data:
    print(obj)
```
